## 2. Enter Your Results
> After running each model notebook, fill in the accuracy (%) and training time (seconds) below.

```python
# ── FILL IN YOUR RESULTS ─────────────────────────────────────────────────────
results = {
    'Model'        : ['Naïve Bayes', 'SVM (RBF)', 'Decision Tree', 'CNN'],
    'Accuracy (%)'  : [0.0, 0.0, 0.0, 0.0],  # Replace with your values
    'Time (s)'     : [0.0, 0.0, 0.0, 0.0],   # Replace with your values
    'Rows Used'    : ['1.44M', '400K', '1.44M', '1.44M'],
    'Type'         : ['ML', 'ML', 'ML', 'DL'],
    'Interpretable': ['High', 'Medium', 'Very High', 'Low'],
}
# ─────────────────────────────────────────────────────────────────────────────

df_results = pd.DataFrame(results)
df_results = df_results.sort_values('Accuracy (%)', ascending=False).reset_index(drop=True)
print('Results Table:')
df_results
```

## 3. Accuracy Comparison Bar Chart

```python
colors = ['#2196F3', '#FF9800', '#4CAF50', '#9C27B0']
models = df_results['Model']
accs   = df_results['Accuracy (%)']

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(models, accs, color=colors, edgecolor='white', linewidth=1.5, width=0.5)

for bar, val in zip(bars, accs):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

ax.set_title('IoT-NetGuard — Model Accuracy Comparison', fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_ylim(0, 110)
ax.yaxis.grid(True, alpha=0.3)
ax.set_axisbelow(True)
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('outputs/comparison_accuracy.png', dpi=150)
plt.show()
print('Saved → outputs/comparison_accuracy.png')
```

## 4. Accuracy vs Training Time Scatter

```python
fig, ax = plt.subplots(figsize=(9, 6))

for i, row in df_results.iterrows():
    ax.scatter(row['Time (s)'], row['Accuracy (%)'],
               s=200, color=colors[i], zorder=5, edgecolors='white', linewidths=1.5)
    ax.annotate(row['Model'],
                (row['Time (s)'], row['Accuracy (%)']),
                textcoords='offset points', xytext=(10, 5), fontsize=11)

ax.set_title('Accuracy vs Training Time (lower-right = best)', fontsize=13, fontweight='bold')
ax.set_xlabel('Training Time (seconds)', fontsize=12)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('outputs/comparison_acc_vs_time.png', dpi=150)
plt.show()
print('Saved → outputs/comparison_acc_vs_time.png')
```
